```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
from scipy import stats
```

```python
x=np.linspace(0,2,100)
y=np.sin(x)
plt.figure()
plt.plot(x,x,'r--',label="x")

plt.title('sin')
plt.show()
```

```python
fig = plt.figure()  # an empty figure with no axes
fig.suptitle('No axes on this figure')  # Add a title so we know which it is
fig, ax_lst = plt.subplots(2,2)

ax_lst[0,0].plot(x, x, label='linear')
ax_lst[0,0].plot(x, x**2, label='quadratic')
ax_lst[0,0].plot(x, x**3,label='cubic')


data={'a':np.arange(50),
      'c':np.random.randint(0,50,50),
      'd':np.random.randn(50)}

data['b']=data['a']+np.random.randn(50)*10
data['d']=np.abs(data['d'])*100

ax_lst[0,1].scatter('a','b',c='c',s='d',data=data)
ax_lst[0,1].set_xlabel('xEntry')
ax_lst[0,1].set_ylabel('yEntry')
```

```python
mu = 100 # mean of distribution
sigma = 15 # standard deviation of distribution
data = mu + sigma * np.random.randn(10000)

num_bins = 50
# the histogram of the data
plt.grid(True)
n, bins, patches = plt.hist(data, num_bins,density=1,facecolor='r', alpha=0.5)
# add a 'best fit' line
y=stats.norm.pdf(bins,mu,sigma)
plt.plot(bins,y)
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
# plt.plot(bins, y, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
```

```python
np.random.seed(19680801)
plt.subplot(221)
y = np.random.normal(loc=0.5,scale=0.4,size=10000)
y = y[(y>0)  & (y<1)]
y.sort()
x = np.arange(len(y))

plt.subplot(221)
plt.plot(x,y)
plt.title('linear')
plt.yscale('linear')

plt.grid(True)
plt.subplot(222)
plt.plot(x,y)
plt.title('log')
plt.yscale('log')
plt.grid(True)
```
